```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/Walmart.csv")
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Week"] = df["Date"].dt.isocalendar().week
```

# for time series the data should be sorted according to date

```python
df = df.sort_values("Date")
```

```python
df[["Date", "Weekly_Sales"]].head()
```

```python
df[["Date", "Weekly_Sales"]].tail()
```

```python
df = df.sort_values("Date").reset_index(drop=True)
```

# creating lag features

```python
df["Lag_1"] = df["Weekly_Sales"].shift(1)
```

```python
df[["Date", "Weekly_Sales", "Lag_1"]].head(10)
```

```python
df["Lag_2"] = df["Weekly_Sales"].shift(2)
df["Lag_4"] = df["Weekly_Sales"].shift(4)
```

```python
df[["Weekly_Sales", "Lag_1", "Lag_2", "Lag_4"]].head(10)
```

```python
df["Rolling_Mean_4"] = df["Weekly_Sales"].rolling(window=4).mean()
```

```python
df[["Weekly_Sales", "Rolling_Mean_4"]].head(10)
```

```python
df["Rolling_Std_4"] = df["Weekly_Sales"].rolling(window=4).std()
```

```python
df[["Weekly_Sales", "Rolling_Mean_4", "Rolling_Std_4"]].head(10)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.to_csv("../data/processed/featured_data.csv", index=False)
```
